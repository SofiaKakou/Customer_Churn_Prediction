# 4.Basic Cleanup

```python
# checking for missing values
df.isna().sum()
```

```python
# checking for duplicates
df.duplicated().sum()
```

```python
# converting "yes/no" to 0/1

for col in df.columns:
  if set(df[col].unique()) == set(['Yes', 'No']):
    df[col] = df[col].map({'Yes':1, 'No':0})
```

```python
# encoding categorical features

df_encoded = pd.get_dummies(df, drop_first=True)
df_encoded.head()
```

```python
# scale numerical features

scaler = StandardScaler()

num_cols = df_encoded.select_dtypes(include=["int64", "float64"]).columns
df_encoded[num_cols] = scaler.fit_transform(df_encoded[num_cols])

df_encoded.head()
```

# 5.Save Clean Dataset

```python
save_path="/content/drive/MyDrive/Thesis/Datasets/telecom_cleaned.csv"
df_encoded.to_csv(save_path, index=False)
```
